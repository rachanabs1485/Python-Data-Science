```python
#DATA SCIENCE SESSION 21 - 1ST MARCH

#Clean the data and see if u have any null or duplicate values



from google.colab import files
uploaded = files.upload()

import pandas as pd
df = pd.read_csv("student_performance_updated_1000 1.csv")
df.head()

print("Dataset Shape:", df.shape)
print (df.head())

df = df.drop(columns=["StudentID", "Name"])

print ("\nMissing Values Per Column:")
print(df.isnull().sum())

duplicates = df.duplicated().sum()
print ("InDuplicate Rows:", duplicates)

df = df.drop_duplicates()

print ("Shape After Removing Duplicates:", df.shape)


#filling the missing values
for col in df.columns :
  if df[col].dtype in ['int64', 'float64']:
    med_val = df[col].median()
    df[col] = df[col].fillna(med_val)
  else:
    mode_val = df[col].mode()[0]
    df[col] = df[col].fillna(mode_val)

print("\nMissing Values Per Column:")
print(df.isnull().sum())


# Gender → Binary Encoding
df ["Gender"] = df[ "Gender"]. map({"Male": 0, "Female": 1})

df["Online Classes Taken"] = df["Online Classes Taken"].astype(int)

df[ "ParentalSupport"] = df["ParentalSupport"]. map({
"LOW": 0,
"Medium": 1,
"High": 2
})


def categorize(score) :
   if score < 60:
     return 0 #Low
   elif score < 80:
     return 1 # Medium
   else:
     return 2 # High

df["Performance_Level"] = df["FinalGrade"].apply (categorize)
```

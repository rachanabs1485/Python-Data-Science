```python
#DATA SCIENCE SESSION 06 - 10TH FEB

student = {
    "Aman": [80,90,70,60],
    "Riya": [95,92,88,91],
    "Sam": [40,55,60,30]
}

topper = ''
top_average = 0
failed_students = [] # Corrected syntax for initializing an empty list and renamed for clarity

for name, marks in student.items():
  average = sum(marks) / len(marks) # Fixed typo: maeks -> marks
  grade = '' # Initialize grade for each student
  if average >= 90:
    grade = 'A'
  elif average >= 75:
    grade = 'B'
  elif average >= 60:
    grade = 'C'
  else: # Corrected logic to assign 'Fail' only if average is below 60
    grade = 'Fail'
    failed_students.append(name) # Append student to failed list only if they fail

  print(f"{name}'s average: {average:.2f}, Grade: {grade}") # Corrected print statement

  if average > top_average: # Changed >= to > to handle cases where multiple students have the same top score, picking the first one
    top_average = average # Corrected assignment operator
    topper = name

# Print topper and failed students outside the loop
print("\n--- Summary ---")
print("Topper :", topper, f"with an average of {top_average:.2f}")
print("Failed students :", failed_students)



account = {
    "Rachana" : 1000
}

while True :
  choice = input("type - deposit/ withdraw/check/exit :")
  if choice == 'deposit' :
    deposit_amount = int(input("Enter amount to deposit:"))
    account["Rachana"] = account['Rachana'] + deposit_amount
    print(f"Deposit successful. New balance: {account['Rachana']}")
  elif choice == 'withdraw':
    withdraw_amount = int(input("Enter amount to withdraw:"))
    if withdraw_amount <= account['Rachana']:
        account['Rachana'] = account['Rachana'] - withdraw_amount
        print(f"Withdrawal successful. New balance: {account['Rachana']}")
    else:
        print("Insufficient funds.")
  elif choice == 'check':
    print(f"Current balance for Rachana: {account['Rachana']}")
  elif choice == 'exit':
    print("Exiting banking application. Goodbye!")
    break
  else:
    print("Invalid choice, please try again.")
```
